```python
#1st importing numpy array and pandas dataframe
import numpy as np
import pandas as pd
```

```python
#load dataset
dataset= pd.read_csv("/content/snails_gc.csv")
```

```python
#showing the dataset, here on the dataset i modify it manually, 1st i move the sex coloum as the last coulmn.
# and i manually change the gender to numeric from catagory. i assume here that 0 is Male, and 1 is Female.
dataset
```

```python
# here in x i take all the feature with out last column sex which is my target.
# and in y i only take the target (Sex) column
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
#here i split my dataset into training and test size, here i put 30% data for training and 70% for testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=30)
```

```python
# here i am taking logistic regression
#as it is a binary classification. and here i am classyfing binary data sex which include male and female
#than i am here fitting the training datas in my model
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
#now here i am predicting with my testing data
y_pred=logreg.predict(X_test)
```

```python
#here i am showing the accurecy score of my model
from sklearn import metrics
accurecy= metrics.accuracy_score(y_test,y_pred)
print("accurecy score:- ", accurecy)
```
